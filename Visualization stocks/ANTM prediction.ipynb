```python
import pandas as pd
import os
import numpy as np
import mplfinance as mpf
from scipy import stats
import plotly.graph_objects as go
os.chdir("D:/Karir/Evidensi/stocks")
```

```python
antm = pd.read_csv('ANTM.JK.csv')
antm = antm.sort_values('Date')
```

```python
antm.isnull().sum()
```

```python
df = antm.set_index(pd.DatetimeIndex(antm['Date'].values))
df
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")
figure = go.Figure(
    data = [
        go.Candlestick(
            x = df.index,
            low = df['Low'],
            high = df['High'],
            close = df['Close'],
            open = df['Open'],
            increasing_line_color = 'green',
            decreasing_line_color = 'red'
        )
    ]
)
figure.update_layout(xaxis_rangeslider_visible = True)

figure.show()
```
